```python
import random
# In below function it writes a list to a file by first writing list = [, and then subsequently writing random integers
# to the file and putting comma between until the last number is reached. Finally writing the ].
def create(n,a,b,filename):
    f = open(filename, 'w+')
    f.write('list = [')
    while n > 0:
        f.write(str(random.randint(a,b)))
        n = n - 1
        if n != 0:
            f.write(',')
        
    f.write(']')
    f.close()

create(10,1,30,'list1.txt')
create(15,1,30,'list2.txt')
```

```python
import math

class Rational:
    p = None
    q = None
    # (a) Create a class Rational which is initialized by two integers, p and q, the nominator and denominator

    # Here the self parameter is used to represent properties which belong to class.
    def __init__(self,p,q):
        self.p = p
        self.q = q

    # (b) Add a method to print the rational number as p/q (the __str__ or __repr__ method is useful).
    def __str__(self):
        return f'{self.p}/{self.q}'

    # (c) We would like to represent 10/20 by 1/2 instead, hence write a function that computes the greatest
    # common divisor, and ensure that every rational number is simplified.

    # Here I used greates common divisor of math library to divide them into it for simplified version of
    # rationals.
    def print_rationals_with_gcd(self):
        gcd = math.gcd(self.p,self.q)
        return f'{int(self.p / gcd)}/{int(self.q / gcd)}'

    # (d) Add a method so that we can add two rational numbers with r1 + r2, here the __add__() method
    # is useful.
    def __add__(self,r1,r2):
        return r1 + r2

    # (e) Add a method to subtract two rational numbers. (__sub__)
    def __sub__(self,r1,r2):
        return r1 - r2
    
    # (f) Add a method to multiply two rational numbers. (__mul__)
    def __mul__(self,r1,r2):
        return r1 * r2
    
    # (g) Add a method to divide two rational numbers. (__div__)
    def __div__(self,r1,r2):
        return r1 / r2
    
    # (h) Add a method that compares whether two rational numbers are equal.
    def compare(self,r1,r2):
        return r1 == r2
    
    # (i) Add a method to convert the rational number to a floating point (the __float__() method maybe handy).
    def __float__(self,r1):
        return r1 / 1.0

    # (j) Add any more functionality that you think is useful but I failed to mention.
    # I also couldn't fide any more.

r1 = Rational(4,6)
print(r1)
print(r1.print_rationals_with_gcd())
print(r1.__add__(1/3,2/7))
print(r1.__sub__(1/3,2/7))
print(r1.__mul__(1/3,2/7))
print(r1.__div__(1/3,2/7))
print(r1.compare(1/3,2/7))
print(r1.__float__(2/7))
```

```python
class Game:

    def __init__(self, agent1, agent2):
        self.moves_a1 = []
        self.moves_a2 = []
        self.nround = 0
        self.score_a1 = 0
        self.score_a2 = 0

        self.agent1 = agent1
        self.agent2 = agent2

    def round(self):
        print(f'Round {self.nround+1}: ')
        # we pass along history so a player can take that into account when
        # deciding the next move
        move_a1 = self.agent1.getMove(self.moves_a2, self.moves_a1)
        # In below line I changed the parameter of moves history of both agent to move of other agent to use in getMove of 
        # MyAgent class.
        move_a2 = self.agent2.getMove(move_a1)

        # compare the moves and decide who wins
        outcome = self.compare(move_a1, move_a2)

        # output outcome of current round
        if outcome == 1:
            print(f'Player 1 wins: {move_a1} beats {move_a2}')
        elif outcome == 0:
            print(f'This round is tied, both players played {move_a1}')
        else:
            print(f'Player 2 wins: {move_a2} beats {move_a1}')

        # updating scores based on the outcome of current round. If outcome is 0 which means tie, then we dont
        # have to do anything since there will be no effect.
        if outcome == 1:
            self.score_a1 = self.score_a1 + 1
        elif outcome == -1:
            self.score_a2 = self.score_a2 + 1
        
        self.moves_a1.append(move_a1)
        self.moves_a2.append(move_a2)

    def play(self, nrounds):
        for r in range(nrounds):
            self.round()
            self.nround += 1

    def summary(self):
        # print some summary of the games
        print('='*20)
        if self.score_a1 > self.score_a2:
            print('Player 1 wins')
        elif self.score_a1 < self.score_a2:
            print('Player 2 wins')
        else:
            print('The game ends in a tie')

        print(f'Final score: {self.score_a1} - {self.score_a2}')

        # Finding the move that was played most often for the both agents.
        num_of_rock = 0
        num_of_paper = 0
        num_of_scissors = 0

        # Here the for loop and if-else statments check each move in both move of a1 and a2, increment the
        # corresponding num variable of moves.
        for i in range(len(self.moves_a1)):
            if self.moves_a1[i] == 'r':
                if self.moves_a2[i] == 'r':
                    num_of_rock = num_of_rock + 2
                elif self.moves_a2[i] == 'p':
                    num_of_rock = num_of_rock + 1
                    num_of_paper = num_of_paper + 1
                elif self.moves_a2[i] == 's':
                    num_of_rock = num_of_rock + 1
                    num_of_scissors = num_of_scissors + 1
            elif self.moves_a1[i] == 'p':
                if self.moves_a2[i] == 'p':
                    num_of_paper = num_of_paper + 2
                elif self.moves_a2[i] == 's':
                    num_of_paper = num_of_paper + 1
                    num_of_scissors = num_of_scissors + 1
                elif self.moves_a2[i] == 'r':
                    num_of_paper = num_of_paper + 1
                    num_of_rock = num_of_rock + 1
            elif self.moves_a1[i] == 's':
                if self.moves_a2[i] == 's':
                    num_of_scissors = num_of_scissors + 2
                elif self.moves_a2[i] == 'r':
                    num_of_scissors = num_of_scissors + 1
                    num_of_rock = num_of_rock + 1
                elif self.moves_a2[i] == 'p':
                    num_of_scissors = num_of_scissors + 1
                    num_of_paper = num_of_paper + 1

        # Here the max() function finds the biggest number of moves among each other.
        most_played_move_num = max(num_of_rock,num_of_paper,num_of_scissors)
        most_played_move = ''

        # And here below if-else statements finds the most played move by checking their numbers.
        if num_of_rock == most_played_move_num:
            most_played_move = 'r'
        elif num_of_paper == most_played_move_num:
            most_played_move = 'p'
        elif num_of_scissors == most_played_move_num:
            most_played_move = 's'
        elif (num_of_rock + num_of_paper) == (2 * most_played_move_num):
            most_played_move = 'r and p'
        elif (num_of_paper + num_of_scissors) == (2 * most_played_move_num):
            most_played_move = 'p and s'
        elif (num_of_scissors + num_of_rock) == (2 * most_played_move_num):
            most_played_move = 's and r'
        else:
            most_played_move = 'r and p and s'
        print(f'Most played move: {most_played_move}')
        print('='*20)

    def compare(self, move1, move2):
        # returns 1 if move1 wins, 0 if it's a tie, and -1 if move2 wins
  
        # conditions which move1 wins
        if (move1 == 'r') & (move2 == 's'):
            return 1
        elif (move1 == 'p') & (move2 == 'r'):
            return 1
        elif (move1 == 's') & (move2 == 'p'):
            return 1
        # conditions which move2 wins
        elif (move2 == 'r') & (move1 == 's'):
            return -1
        elif (move2 == 'p') & (move1 == 'r'):
            return -1
        elif (move2 == 's') & (move1 == 'p'):
            return -1
        # tie
        return 0
```

### In exercise 6.5 you implemented functions for sparse and dense vector multiplications using lists and dictionaries. However, this is a bit clumsy to use in practice.
